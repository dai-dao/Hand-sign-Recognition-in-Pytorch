### PyTorch Segmentation Results on Validation set

1. Results suggests that training should focus more on lighting augmentation. The model 
    performs well in shadows, however perform poorly on too bright settings. 
    => Increase bright light augmentation during training.
2. In normal cases, the hand is perfectly segmented, which enables contours finding to be easy 
     which in turn helps determine the number of fingers.
3. Faraway fingers appear very blurry, due to small size and shadow. Might be able to overcome
    when trained with more light augmentation.

```python
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.01, hspace=0.02) # set the spacing between axes.
plt.figure(figsize=(12,12))

for i in range(64):
    ax1 = plt.subplot(gs1[i])
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])
    ax1.set_aspect('equal')
    
    plt.subplot(8, 8, i+1)
    plt.imshow(norm_pytorch[i][:, :, 0])
    plt.axis('off')

plt.show()
```
